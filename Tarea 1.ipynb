```python
# Importamos bibliotecas necesarias 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Valores entregados por el problema
H_thin = 350  # altura escalar del disco delgado en PC
H_thick = 1000  # altura escalar del disco grueso en PC
h_r = 2250  # escala radial en PC
n_0 = 1 # Asumiremos 

# Definimos la función de densidad estelar n(z,R) para el disco delgado
def n_thin(z, R):
    """Función que calcula la densidad estelar para el disco delgado de la Galaxia."""
    thin_term = np.exp(-z / H_thin)
    return thin_term

# Definimos la función de densidad estelar n(z,R) para el disco grueso
def n_thick(z, R):
    """Función que calcula la densidad estelar para el disco grueso de la Galaxia."""
    thick_term = 0.085 * np.exp(-z / H_thick)
    return thick_term

# Creamos datos para la visualización
z_values = np.linspace(-3000, 3000, 100)
R_values = np.linspace(0, 10000, 100)
Z, R = np.meshgrid(z_values, R_values)

# Calculamos la densidad estelar para el disco delgado y grueso
n_thin_values = n_thin(Z, R)
n_thick_values = n_thick(Z, R)

# Sumamos las contribuciones del disco delgado y grueso para obtener la densidad total y multiplicamos por el decaimiento
n_total_values = (n_thin_values + n_thick_values) * np.exp(-R / h_r)


# Graficando en 3D
fig = plt.figure(figsize=(9, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(Z, R, n_total_values, cmap='viridis', edgecolor='pink')
ax.set_xlabel('Z (pc)')
ax.set_ylabel('R (pc)')
ax.set_zlabel('Densidad estelar')
ax.set_title('Distribución 3D del Disco Galáctico')

plt.show()
```
